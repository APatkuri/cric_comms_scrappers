```python
# List of desired columns
columns_to_select = [
    'p_match', 'inns', 'bat', 'p_bat', 'team_bat', 'bowl', 'p_bowl', 'team_bowl',
    'ball', 'ball_id', 'outcome', 'score', 'out', 'dismissal', 'p_out', 'over',
    'noball', 'wide', 'byes', 'legbyes', 'cur_bat_runs', 'cur_bat_bf', 'cur_bowl_ovr',
    'cur_bowl_wkts', 'cur_bowl_runs', 'inns_runs', 'inns_wkts', 'inns_balls', 'trail_by',
    'lead_by', 'day', 'session', 'target', 'date', 'year', 'ground', 'country',
    'winner', 'toss', 'competition', 'bat_hand', 'bowl_style', 'bowl_kind', 'batruns',
    'ballfaced', 'bowlruns', 'bat_out', 'wagonX', 'wagonY', 'wagonZone', 'line',
    'length', 'shot', 'control', 'predscore', 'wprob'
]

# Create the new DataFrame with only the specified columns
final_df = pd.DataFrame(columns=columns_to_select)
```

```python
# Mapping of new_df columns to old_df columns
column_mapping = {
    'inns': 'inningNumber',
    'p_bat': 'batsmanPlayerId',
    'p_bowl': 'bowlerPlayerId',
    'ball': 'ballNumber',
    'ball_id': 'oversUnique',
    'score': 'totalRuns',
    'p_out': 'outPlayerId',
    'over': 'overNumber',
    'noball': 'noballs',
    'wide': 'wides',
    'byes': 'byes',
    'legbyes': 'legbyes',

    'cur_bat_runs': '',
    'cur_bat_bf': '',
    'cur_bowl_ovr': '',
    'cur_bowl_runs': '',

    'inns_runs': 'totalInningRuns',
    'inns_wkts': 'totalInningWickets',
    'batsruns': 'batsmanRuns',

    'wagonX':'wagonX',
    'wagonY':'wagonY',
    'wagonZone':'wagonZone',
    'line':'pitchLine',
    'length':'pitchLength',
    'shot':'shotType',
    'control':'shotControl',
    
}

# Iterate through the mapping and assign values
for new_col, old_col in column_mapping.items():
    if old_col in df_updated.columns:
        # Assign values from old_df[old_col] to new_df[new_col]
        final_df[new_col] = df_updated[old_col]
    else:
        # If old_col doesn't exist, fill with NaN or default value
        final_df[new_col] = None  # or np.nan
```

```python
import requests

def get_redirected_commentary_url(match_id):
    # Fetch the redirected URL

    original_url = f"https://www.espncricinfo.com/matches/engine/match/{match_id}.html"
    response = requests.head(original_url, allow_redirects=True)
    redirected_url = response.url
    
    # Replace 'full-scorecard' with 'ball-by-ball-commentary'
    if "full-scorecard" in redirected_url:
        return redirected_url.replace("full-scorecard", "ball-by-ball-commentary")
    else:
        raise ValueError("Redirected URL does not contain 'full-scorecard'. Cannot modify to commentary URL.")

# Example usage:

# commentary_url = get_redirected_commentary_url(1428554)
# print(commentary_url)
```
